Histogram:

```python
import numpy as np
import matplotlib.pyplot as plt
import random
plt.rcParams['figure.dpi'] = 150
plt.rcParams['savefig.dpi'] = 150
x0=0
D=0.5   #step length
N=20000
sigma=1     #STD in Normal Dist.
def P(x):
    return np.exp(-(x**2)/(2*sigma**2))/(sigma*np.sqrt(2*np.pi))
X=np.array([x0])
Y=np.array([x0])
for i in range(N):
    rnd=np.random.uniform(-1,1)
    #print(rnd)
    y=X[i]+(D*rnd)
    #print(X[i],P(X[i]),y,P(y))
    mrnd=random.random()
    r=min(1,P(y)/P(X[i]))
    if mrnd<r:
        X=np.append(X,y)
        Y=np.append(Y,y)
    else:
        X=np.append(X,X[i])
win=(len(Y)-1)/N
print(f'Win Rate for Step Length={D} is {win}')
bin,edges,z=plt.hist(Y,bins=20,label=f'Histogram of {N} data with Metropolis Method')
std=np.std(Y)
mean=np.average(Y)
xx=np.linspace(min(Y),max(Y),100)
yy=np.exp(-0.5*((xx-mean)/std)**2)*max(bin)
plt.plot(xx,yy,'r',label='Normal Dist.')
plt.title(f'Histogram of {N} data with Metropolis Method Shown with a Normal Dist.')
#plt.legend(loc='upper right')
plt.show()
```

Win Rate with Step Length

```python
import numpy as np
import matplotlib.pyplot as plt
import random
plt.rcParams['figure.dpi'] = 150
plt.rcParams['savefig.dpi'] = 150
x0=0
D=0.1   #step length
N=10000
sigma=1     #STD in Normal Dist.
def P(x):
    return np.exp(-(x**2)/(2*sigma**2))/(sigma*np.sqrt(2*np.pi))
steps=np.array([])
WinRate=np.array([])
while True:
    X=np.array([x0])
    Y=np.array([x0])
    for i in range(N):
        rnd=np.random.uniform(-1,1)
        #print(rnd)
        y=X[i]+(D*rnd)
        #print(X[i],P(X[i]),y,P(y))
        mrnd=random.random()
        r=min(1,P(y)/P(X[i]))
        if mrnd<r:
            X=np.append(X,y)
            Y=np.append(Y,y)
        else:
            X=np.append(X,X[i])
    win=(len(Y)-1)/N
    if win<0.09:
        break
    WinRate=np.append(WinRate,win)
    steps=np.append(steps,D)
    D+=0.1
for k in range(len(steps)):
    print(f'w: {WinRate[k]} , s: {steps[k]}')

plt.plot(steps,WinRate,'bo',markersize='4')
plt.title(f'Graph of Win Rate vs. Step Size')
plt.xlabel("Step Size")
plt.ylabel("Win Rate")
plt.show()
```

Correlation length  $\xi$

$C(j) \propto e^{-\frac{j}{\xi}} \Rightarrow \log[C(j)] \propto -\frac{j}{\xi} \Rightarrow \frac{-1}{\text{slope}} = \xi$

```python
import numpy as np
import matplotlib.pyplot as plt
import random
plt.rcParams['figure.dpi'] = 150
plt.rcParams['savefig.dpi'] = 150
x0=0
D=0.1   #step length
N=10000
Js=50
sigma=1     #STD in Normal Dist.
def P(x):
    return np.exp(-(x**2)/(2*sigma**2))/(sigma*np.sqrt(2*np.pi))
WRs=np.array([0.5,1.0,1.6,2.2,2.9,3.9,5.3,7.9,15.1])
steps=np.array([])
WinRate=np.array([])
CorrLen=np.array([])
for D in [0.5]:
    X=np.array([x0])
    Y=np.array([x0])
    for i in range(N):
        rnd=np.random.uniform(-1,1)
        #print(rnd)
        y=X[i]+(D*rnd)
        #print(X[i],P(X[i]),y,P(y))
        mrnd=random.random()
        r=min(1,P(y)/P(X[i]))
        if mrnd<r:
            X=np.append(X,y)
            Y=np.append(Y,y)
        else:
            X=np.append(X,X[i])
    win=(len(Y)-1)/N
    WinRate=np.append(WinRate,win)
    steps=np.append(steps,D)
    CJ=np.array([])
    J=np.array([])
    for j in range(Js):
        C1=np.array([])
        for i in range(len(Y)):
            if (i+j)<len(Y):
                c1=Y[i]*Y[(i+j)]
                C1=np.append(C1,c1)
        J=np.append(J,j)
        cij=np.average(C1)
        mean=np.average(Y)
        std=np.std(Y)
        Cj=(cij-mean**2)/(std**2)
        CJ=np.append(CJ,Cj)
    logCJ=np.log(CJ)
    plt.plot(J,logCJ,'bo',markersize='5')
    #plt.plot(J,CJ,'ro',markersize='5')
    m,b=np.polyfit(J, logCJ, 1)
    corrlen=-1/m
    CorrLen=np.append(CorrLen,corrlen)
print(f'Correlation Lengths are {CorrLen}')
plt.title(f'log C(j) vs. j')
plt.xlabel("j")
plt.ylabel("log C(j)")
plt.show()
```
